Ритейл-задача: отличается ли цена на колготки у Голден Леди и Леванте. Если да, то на сколько?

```python
import pandas as pd
import numpy as np

import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = pd.read_csv('sprint11.csv')
```

изучаем

```python
df.describe()
```

```python
df.pivot_table(index='firm', values='Price', aggfunc='mean')
```

наша первая модель

```python
formula0 = 'Price ~ DEN + polyamid + lykra + cotton + wool'
reg0 = smf.ols(formula0, df).fit()
reg0.summary()
```

используем состоятельные в условиях гетероскедастичности стандартные ошибки

```python
formula0 = 'Price ~ DEN + polyamid + lykra + cotton + wool'
reg0 = smf.ols(formula0, df).fit(cov_type='HC1')
reg0.summary()
```

боремся с мультиколлинеарностью

```python
var = reg0.model.exog
VIF = [variance_inflation_factor(var, i) for i in range(var.shape[1])]
VIF 
```
